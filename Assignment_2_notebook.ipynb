```python
customers_df[['NumWebVisitsMonth', 'NumWebPurchases']].corr()
```

```python
customers_df[['Income', 'NumWebPurchases']].corr()
```

However, at least for the customers around the median of site visits, there is an apparent increase in bubble size (income) as the number of online purchases increases. Furthermore, this visual analysis is confirmed by the **0.39** correlation between income and online purchases, which indicates a weak positive correlation.

**Are campaign responders visually distinguishable? If so, how are they different?**
<br>
Campaign responders are *not* that visually distinguishable. It is apparent that a fair portion of the non-responders are closer toward the little to no site vists and/or little to no online purchases. However, there are also a number of non-responders scattered with higher site visits and/or higher online purchases. The first stated observation makes sense because campaign responders are more likely to visit the website, as well as make purchases online. In this part, we will study customers' pattern/distribution of spendings across different types of customers. We will visualize such patterns for different customer groups using **bar charts**.

### Task B.1 (15 points)

i. Consider that customers are grouped by their **education levels**. Develop a bar chart with several sets of bars, in which:

* Each set of bars (i.e.,bar of the same color) represent the spendings by customers in different groups on a spefic product category. For example, spending by customers "Graduation", "Master", "PhD", ..., on *fish products*.  
* The chart should allow you to easily visuallize both the **total spending** by customers in a group on all product categories and the **composition** of their spending on those product categories.

ii. Further, answer the following questions in a **text (markdown) cell** and make sure to explain your answer based on the chart:
* Which customer group contributes the highest revenue?

**Hint**: By default, each row in the dataset is shown as a separate rectangle in the bar, and these rectangles have borders that can make the chart look cluttered. In that case, use `fig.update_traces(opacity=1, marker_line_width=0)` to remove the border lines.

```python
spendingLong1 = pd.melt(
    customers_df,
    id_vars = 'Education',
    value_vars = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds'],
    var_name = 'Product',
    value_name = 'Amount Spent'
)

figB1 = px.bar(
    spendingLong1,
    x = 'Product',
    y = 'Amount Spent',
    color = 'Education',
    title = 'Customers\' Distribution of Spending Across Different Types of Products Based on Education'
)

figB1.update_layout(
    xaxis_title = 'Product',
    yaxis_title = 'Amount Spent on Product in the Last Two Years',
    legend_title = 'Education',
    width = 1000, height = 900,
    title_x = 0.5
)

figB1.update_traces(opacity=1, marker_line_width=0)

figB1.show()
```

```python
spendingDistribution1 = spendingLong1.groupby(['Product', 'Education'], as_index=False)['Amount Spent'].sum()
spendingDistribution1['Product Total'] = spendingDistribution1.groupby('Product')['Amount Spent'].transform('sum')
spendingDistribution1['Percentage (%)'] = spendingDistribution1['Amount Spent'] / spendingDistribution1['Product Total'] * 100
spendingDistribution1
```

**Which customer group contributes the highest revenue?**
<br>
*Graduation* customers contribute the highest to revenue overall, as well as individually across all product categories.
<ul>
  <li>They account for 47% of the total wine sales, while the next top spending group only accounts for 29%.</li>
  <li>They account for 59% of the total fruit sales, while the next top spending group only accounts for 17%.</li>
  <li>They account for 54% of the total meat product sales, while the next top spending group only accounts for 22%.</li>
  <li>They account for 58% of the total fish product sales, while the next top spending group only accounts for 16%.</li>
  <li>They account for 58% of the total sweet product sales, while the next top spending group only accounts for 16%.</li>
  <li>They account for 58% of the total gold product sales, while the next top spending group only accounts for 16%.</li>
</ul>

```python
figB2a = px.bar(
    spendingLong2,
    x = 'Marital_Status',
    y = 'Amount Spent',
    color = 'Product',
    title = 'Customers\' Distribution of Spending By Marital Status Across Different Types of Products'
)

figB2a.update_layout(
    xaxis_title = 'Marital Status',
    yaxis_title = 'Amount Spent by Customer Marital Status Groups in the Last Two Years',
    legend_title = 'Product Category',
    width = 1200, height = 900,
    title_x = 0.5
)

figB2a.update_traces(opacity=1, marker_line_width=0)

figB2a.show()
```

  <li>Divorced: Wine products account for 53% of their total spending, while the next highest product category only accounts for 25%.</li>
  <li>Widow: Wine products account for 50% of their total spending, while the next highest product category only accounts for 25%.</li>

In this part, let us investigate variability in customers' online purchasing behavior.

i. We will use a visualization chart that shows distribution of customers' online purchases. The chart should:
* Show multiple distributions, each being the distribution of number of online purchases by customers of a certain generation (Gen X vs. Millennials,  ignore other generations)
* Allow users to observe the 1st, 2nd, and 3rd quartiles of the distribution, along with any outliers.

ii. Based on the chart, answer the following question:
* How would you comment on the pattern of online purchases across different generations.

iii. In your chart, further divide each distribution into subgroups of customers based on the **number of kids at home** (not including teens).
How would you comment on the pattern of online purchases across the different subgroups of customers?




```python
customers_df.loc[(customers_df['Year_Birth'] >= 1965) & (customers_df['Year_Birth'] <= 1980), 'Generation'] = 'Gen X'
customers_df.loc[(customers_df['Year_Birth'] >= 1981) & (customers_df['Year_Birth'] <= 1996), 'Generation'] = 'Millennials'

figC1 = px.box(
    customers_df.loc[customers_df['Generation'].notnull()],
    x = 'Generation',
    y = 'NumWebPurchases',
    title = 'Customers\' Distribution of Online Purchase Based on Generation'
)

figC1.update_layout(
    width = 600, height = 800,
    title_x = 0.5
)

figC1.show()
```

**How would you comment on the pattern of online purchases across different generations?**
<br>
Gen X customers appear to make *slightly more* online purchases than millennial customers. Though they both have the same median number of online purchases at 3, the 75th-percentile Gen X customer made 6 online purchases, while the corresponding millennial customer only made 5 online purchases. This is not really much of a meaningful difference, but it is rather surprising because I would have expected millennial customers to be more accustomed to the practice of online shopping than Gen X customers. It is possible that maybe Nata's customer base consists of a more technologically adaptable set of Gen X customers who have transitioned into online shopping.

```python
figC2 = px.box(
    customers_df.loc[customers_df['Generation'].notnull()],
    x = 'Generation',
    y = 'NumWebPurchases',
    color = 'Kidhome',
    labels = {'Kidhome': 'Number of Kids in the Household'},
    title = 'Customers\' Distribution of Online Purchase Based on Generation and Household Kids'
)

figC2.update_layout(
    width = 1000, height = 800,
    title_x = 0.5
)

figC2.show()
```

**How would you comment on the pattern of online purchases across the different subgroups of customers?**
<br>
Customers from households without kids appear to make *slightly more* online purchases than households with 1-2 kids. While the median Gen X and millennial customer from 0-kid households made 5 and 4 online purchases respectively, the median customers from both generations with 1-2 kids made only 2 online purchases. This is rather surprising because I would have expected larger households, especially those with kids, to require a higher quantity of supermarket products, therefore to make more online purchases. However, this does not appear to be the case, so maybe families with kids prefer to make these sort of purchases at brick and mortar stores, and therefore their online purchases are slightly lower.
